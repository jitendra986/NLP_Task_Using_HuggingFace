Hi In this lesson you will measure sentence similarity using the sentence transformer library. Sentence similarity measure how close tow pieces of text for instance the pharese ' I like kittens' and ' We love cats' Have similar meaning.The sentence similarity is specially useful for information retrieval and clustering and grouping

```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentences1 = ['The cat sits outside',
              'A man is playing guitar',
              'The movies are awesome']
```

```python
embeddings1 = model.encode(sentences1, convert_to_tensor=True)
```

```python
embeddings1
```

```python
sentences2 = ['The dog plays in the garden',
              'A woman watch TV',
              'The new movie is so great']
```

```python
embeddings2 = model.encode(sentences2, convert_to_tensor=True)
```

```python
embeddings2
```

```python
sentences3 = [
    "This framework generates embeddings for each input sentence",
    "Sentences are passed as a list of string.",
    "The quick brown fox jumps over the lazy dog.",
]
embeddings3 = model.encode(sentences3, convert_to_tensor=True)
```

```python
embeddings3.shape
```

We can calculate how close these sentences are to do that we use the cosine distences which is a measure to calculate how close and how far two vectors are?
To do that we need 'util' function

```python
from sentence_transformers import util
```

```python
cosine_scores = util.cos_sim(embeddings1,embeddings2)
```

```python
print(cosine_scores)
```

notice that these are pairwise similarity for every sentence in the first list to every sentence to the second list.Now lets output the score for each pair of both list

```python
for i in range(len(sentences1)):
  print("{} \t\t {} \t\t Score: {:.4f}".format(sentences1[i],
                                               sentences2[i],
                                               cosine_scores[i][i]))
```

Let's build a classification pipeline. For this kind of audio classification. you will need a pretrain clap model.at a moment it one of a kind architecture that is available for these task, you can find it on hugging facehub by filtering model with feature extraction multimodel task and then filtering it by name 'clap'. To classify your audio example you only need the array of audio data, However the example has to have a sampling rate that the model expects. Lets talk about sampling rate. a sound wave is a continous signal this means it contains an infinite number of signal values in a given time. but the audio your computer can work with is a series of discrete values known as 'digital representation'. To get the 'digital representation' of a contionous audio signal.we first capture the sound with a micro phone then the analog signal is converted into an electrical signal then the electrical signal is sampled to get the digital representation. Sampling means measuring the values of continous signal at fixed time steps as a result the sampled wave form is discrete with finite number of values at uniform intervals. A very important characteristics of digitiged audio is the sampling rate it is the number of samples taken in 1 seconds.and it is measured in (Hz).
8KHz is a sampling rate of an audio in a telephone.
16KHz is a sampling rate is good enough to capture human speech without it souding muffle.
192KHz is a sampling rateof a professional high definition audio recording equipment. But why is the sampling rate is important when working with an AI model.Sampling rate for transformer models

5- Second sounds

#At 8,000 Hz ==> 5*8000 = 40,000 signal values
#At 16,000 Hz ==>5*16000 = 80,000 signal values
#At 192,000 Hz ==>5*192000 = 960,000 signal values

For transformer models these three arrays are very different. Transformer model treat input as sequences and rely on attention mechanism to learn audio representation. They are train on datasets where all examples have the same sampling rate and they dont generalize well to the other sample rate. So for a transformer model train on 16KHz audio a 5 second high quality audio(192KHz audio)  that is expressed with almost 1 million signal values will look like a 16 second recordings.

If a transformer model has been trained with audio samples each recorded at a sampling rate of 16KHz. it is going to view any input as it was recorded at the same sampling rate. So
Lets take a 1 second sound which has been recorded with 192KHz sampling rate, So how many values will the array representing that sound have

```python
1*192000
```

But if we take this same model that has been trained on audio exaples with 16KHz Sampling rate. Is it going to see as 1 seconds or more? Lets's find out

```python
192000/16000
```

So for this model the recording is going to look like 12 seconds recordings.

```python
mic_transcribe = gr.Interface(
    fn = transcribe_speech,
    inputs = gr.Audio(sources='microphone',
                      type = 'filepath'),
    outputs = gr.Textbox(label = 'Transcription',
                         lines=3),
    allow_flagging = 'never')
```



*   Build the Gradio interface.




```python
import gradio as gr
demo = gr.Blocks()
```

```python
def transcribe_long_form(filepath):
    if filepath is None:
        gr.Warning("No audio found, please retry.")
        return ""
    output = asr(
      filepath,
      max_new_tokens=256,
      chunk_length_s=30,
      batch_size=8,
    )
    return output["text"]
```

```python
mic_transcribe = gr.Interface(
    fn=transcribe_long_form,
    inputs=gr.Audio(sources="microphone",
                    type="filepath"),
    outputs=gr.Textbox(label="Transcription",
                       lines=3),
    allow_flagging="never")

file_transcribe = gr.Interface(
    fn=transcribe_long_form,
    inputs=gr.Audio(sources="upload",
                    type="filepath"),
    outputs=gr.Textbox(label="Transcription",
                       lines=3),
    allow_flagging="never",
)
```

```python
with demo:
    gr.TabbedInterface(
        [mic_transcribe,
         file_transcribe],
        ["Transcribe Microphone",
         "Transcribe Audio File"],
    )
demo.launch(share=True,
            server_port=int(os.environ['PORT1']))
```
